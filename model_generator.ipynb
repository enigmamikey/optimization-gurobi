```python
#define functions

def sortVectorOfVectors(x):
    for i in range(len(x)-2):
        for j in range(1,len(x)-1-i):
            if getFirstVector(x[j],x[j+1]) == 1:
                swap(x,j,j+1)
    return(x)

def getFirstVector(x,y):
    for i in range(len(x)):
        if x[i] > y[i]:
            return 1
        if y[i] > x[i]:
            return 2
        
def swap(x,a,b):
    temp = x[a]
    x[a] = x[b]
    x[b] = temp
    return x

def sort(x):
    for i in range(len(x) - 1):
        maxValue = 0
        for j in range(len(x) - i):
            if x[j] > maxValue:
                maxValue = x[j]
                maxValueIndex = j
        temp = x[len(x) - i - 1]
        x[len(x) - i - 1] = maxValue
        x[maxValueIndex] = temp
    return x

def printTerm(y,z):
    j = 0
    highCount = 3
    for i in range(1,len(y)):
        print('y', end="")
        print(y[i], end="")
        j = j + 1
        if y[i] >= 10:
            highCount = highCount - 1
    for i in range(1,len(z)):
        print('z', end="")
        print(z[i], end="")
        if j == 4:
            print('*', end="")
        j = j + 1
        if z[i] >= 10:
            highCount = highCount - 1
    for i in range(highCount):
        print(' ', end="")
    print(' + ', end="")
        
def setVector(x,ySpot,k):
    y = [0]
    z = [0]
    for i in range (len(x)):
        if i in ySpot:
            if x[i] + k <= 11:
                y = y + [x[i] + k]
            else:
                y = y + [x[i] + k - 11]
        else:
            if x[i] + k <= 11:
                z = z + [x[i] + k]
            else:
                z = z + [x[i] + k - 11]
    y = sort(y)
    z = sort(z)
    printTerm(y,z)
    if k % 4 == 3:
        print('\\')
    return [y + z]
```
